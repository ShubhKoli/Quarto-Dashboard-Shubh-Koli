```python
# Now we know what points to consider we would be making graphs : For the datset : 


# Lets make the Line Graph for the Country, Time Period, 

# sns.lineplot(df_1['time_period'] , df['obs_value'] , data=df_metadata)
sns.lineplot(data=df_1 , x ='time_period' , y='obs_value' )
plt.xlabel("Time Period - From 1950 to 2020")
plt.ylabel("Obs Value  - Old Age Dependency Value")
plt.show()
```

```python
sns.histplot(data=df_1 , x = 'time_period' , y='obs_value' , bins=20  , hue='country' , legend=False)
plt.xlabel("Time Period - From 1950 to 2020")
plt.ylabel("Obs Value  - Old Age Dependency Value")
plt.show()
```

```python
sns.scatterplot(data=df_metadata , x='year' , y='Population, total' , hue="country" , legend=False)
```

```python
plt.figure(figsize=(16 ,10))
sns.countplot(data=df_metadata , x = 'year')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.lineplot(data=df_metadata , x="year" , y='Life expectancy at birth, total (years)')
plt.xlabel("Data - Over the Years")

plt.show()
```
